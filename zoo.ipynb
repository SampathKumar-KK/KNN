## Problem Statement
### Implement a KNN model to classify the animals into categories

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('zoo.csv')

df.head()
```

```python
df[df.duplicated()]
```

```python
df = df.drop(columns=['animal name'])
```

```python
d = {'type': {'1':'Mammal','2':'Bird','3':'Reptile','4':'Fish','5':'Amphibian','6':'Bug','7':'Invertibrate'}}
```

```python
df=df.replace(d)
df.head()
```

```python
import matplotlib.pyplot as plt

df.type.value_counts().plot(kind ='bar')
plt.title('Frequency of Type')
plt.xlabel("Type of Animals")
plt.ylabel("Count")
plt.show()
```

## Data Preprocessing

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
X.head()
```

```python
Y.head()
```

```python
kfold = KFold(n_splits=10)

model = KNeighborsClassifier(n_neighbors= 17)
results = cross_val_score(model, X, Y, cv= kfold)

print(results.mean())
```

```python
"""Grid Search for Algorithm Tuning to find best value of K"""

from sklearn.model_selection import GridSearchCV

n_neighbours = np.array(range(1,40))
param_grid = dict(n_neighbors= n_neighbours)

model = KNeighborsClassifier()
grid = GridSearchCV(estimator = model, param_grid = param_grid)
grid.fit(X,Y)

print(grid.best_score_)
print(grid.best_params_)
```

```python
"""Visualizing the CV results"""


import matplotlib.pyplot as plt

#Choose k between 1 to 41

k_range = range(1,41)
k_scores = []

# Use iteration to calculate different k in models, then return the average accuracy based on the cross Validation

for k in k_range:
  knn = KNeighborsClassifier(n_neighbors= k)
  scores = cross_val_score(knn, X, Y, cv= 5)
  k_scores.append(scores.mean())
```

```python
# plot to see clearly

plt.plot(k_range, k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
pd.DataFrame(k_scores).sort_values(0,ascending=False).head(10)
```
